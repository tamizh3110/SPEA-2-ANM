```julia
function kur_prime(solution_set)
    temp=Matrix(undef,size(solution_set,1),number_of_objectives)
    for i in 1:size(solution_set,1)
        fitness_objective_1=0
        fitness_objective_2=0
        for j in 1:size(solution_set,2)
            fitness_objective_1+=abs(solution_set[i,j])^0.8+5*(sin(solution_set[i,j])^3)+3.5828
        end
        for k in 1:size(solution_set,2)-1
            fitness_objective_2+=(1-exp(-0.2*sqrt(solution_set[i,k]^2+solution_set[i,k+1]^2)))
        end
        temp[i,1],temp[i,2]=fitness_objective_1,fitness_objective_2
    end
    return temp
end

    
    

            
    
```

```julia
function QV(solution_set)
    temp=Matrix(undef,size(solution_set,1),number_of_objectives)
    for i in 1:size(solution_set,1)
        sum_value_one=0
        sum_value_two=0
        for j in 1:size(solution_set,2)
            sum_value_one+=((solution_set[i,j]^2)-10*cos(2*pi*solution_set[i,j])+10)
            sum_value_two+=((solution_set[i,j]-1.5)^2-10*cos(2*pi*(solution_set[i,j]-1.5))+10)
        end
        fitness_objective_1=((1/decision_space_dimensions)*sum_value_one)^(0.25)
        fitness_objective_2=((1/decision_space_dimensions)*sum_value_two)^(0.25)
        temp[i,1],temp[i,2]=[fitness_objective_1,fitness_objective_2]
    end
    return temp
end
            
```

```julia
function kursawe_fitness_function_evaluation(solution_set)
    temp=Matrix(undef,size(solution_set,1),number_of_objectives)
    for i in 1:size(solution_set,1)
        fitness_objective_1=0
        fitness_objective_2=0
        for j in 1:size(solution_set,2)-1
            power_value=-0.2*sqrt(solution_set[i,j]^2+solution_set[i,j+1]^2)
            fitness_objective_1+=-10*exp(power_value)
        end
        for k in 1:size(solution_set,2)
            fitness_objective_2+=(abs(solution_set[i,k]))^0.8+5*sin(solution_set[i,k]^3)
        end
        temp[i,1],temp[i,2]=[fitness_objective_1,fitness_objective_2]
    end
    return temp
end

            
            
```

```julia
function fs_1(solution_set)
    temp=Matrix(undef,size(solution_set,1),number_of_objectives)
    even_set=[2,4,6,8,10]
    odd_set=[3,5,7,9]
    for i in 1:size(solution_set,1)
        fitness_objective_1=0
        fitness_objective_2=0
        sum_val=0
        sum_val_two=0
        for val in odd_set
            sum_val+=(solution_set[i,val]-solution_set[i,1]^(0.5*(1.0+(3*(val-2)/8))))^2
        end
        for val_two in even_set
            sum_val_two+=(solution_set[i,val_two]-solution_set[i,1]^(0.5*(1.0+(3*(val_two-2)/8))))^2
        end
        fitness_objective_1=solution_set[i,1]+(1/2)*sum_val
        fitness_objective_2=1-sqrt(solution_set[i,1])+(2/5)*sum_val_two
        temp[i,1],temp[i,2]=[fitness_objective_1,fitness_objective_2]
    end
    return temp
end

                
    

    
```

```julia
function non_domination_verification(fitness_set,density_set)
#     non dominated setup
    fitness_obj_1=Array{Float64}(undef,size(fitness_set,1))
    fitness_obj_2=Array{Float64}(undef,size(fitness_set,1))
    dominators=[]
    list_of_dominations=[]
    raw_fitness=Array{Float64}(undef,size(fitness_set,1))
    for i in 1:size(fitness_set,1)
        fitness_obj_1[i]=fitness_set[i,1]
        fitness_obj_2[i]=fitness_set[i,2]
    end
    for i in 1:size(fitness_obj_1,1)
        dominator=0
        temp_list=[]
        for j in 1:size(fitness_obj_2,1)
            if (fitness_obj_1[i] < fitness_obj_1[j] && fitness_obj_2[i] < fitness_obj_2[j]) || (fitness_obj_1[i] <= fitness_obj_1[j] && fitness_obj_2[i] < fitness_obj_2[j]) || (fitness_obj_1[i] < fitness_obj_1[j] && fitness_obj_2[i] <= fitness_obj_2[j])
                dominator+=1
            elseif (fitness_obj_1[i] > fitness_obj_1[j] && fitness_obj_2[i] > fitness_obj_2[j]) || (fitness_obj_1[i] >= fitness_obj_1[j] && fitness_obj_2[i] > fitness_obj_2[j]) || (fitness_obj_1[i] > fitness_obj_1[j] && fitness_obj_2[i] >= fitness_obj_2[j])
                push!(temp_list,j)
            end
        end
        push!(dominators,temp_list)
        push!(list_of_dominations,dominator)
    end
    
    #raw_fitness_setup
    raw_fitness_setup=[]
    for (i,val) in enumerate(dominators)
        raw_fitness=0
        if length(val)!=0
            for j in 1:size(val,1)
                raw_fitness+=list_of_dominations[val[j]]
            end
#         println("raw fitness before adding density val",raw_fitness)
            raw_fitness+=density_set[i]
#         println("raw fitness after adding density val",raw_fitness)
            push!(raw_fitness_setup,raw_fitness)
        elseif length(val)==0
             raw_fitness=density_set[i]
             push!(raw_fitness_setup,raw_fitness)
        end
    end
    return raw_fitness_setup
end
 
# non_dom=non_domination_verification(fit,dense)
# println(non_dom)
```

```julia
function polynomial_mutation(mut,lower,upper,eta)
    for i in 1:size(mut,1)
        for j in 1:size(mut,2) 
            if Base.rand()<=prob_of_mutation
                delta_1=(mut[i,j]-lower)/(upper-lower)
                delta_2=(upper-mut[i,j])/(upper-lower)
                mutation_strength=1.0/(1.0+eta)
                random_value=Base.rand()
                if random_value <= 0.5
                    value=2.0*random_value +(1.0-2.0*random_value)*(1.0-delta_1)^(1.0+eta)
                    delta_q=value^mutation_strength-1.0
                else
                    value=2.0*(1.0-random_value)+2.0*(random_value-0.5)*(1.0-delta_2)^(1.0+eta)
                    delta_q=1.0-value^mutation_strength
                end
                mut[i,j]=mut[i,j]+delta_q*(upper-lower)
                mut[i,j]=Base.min(Base.max(mut[i,j],lower),upper)
            end
        end
    end
    return mut
end
        
    
```

```julia
function polynomial_mutation_zdt_4(mut,lower,upper,eta)
    for i in 1:size(mut,1)
        for j in 1:size(mut,2) 
            if Base.rand()<=prob_of_mutation
                if j==1
                    delta_1=mut[i,j]
                    delta_2=1.0-mut[i,j]
                    mutation_strength=1.0/(1.0+eta)
                    random_value=Base.rand()
                    if random_value <= 0.5
                        value=2.0*random_value +(1.0-2.0*random_value)*(1.0-delta_1)^(1.0+eta)
                        delta_q=value^mutation_strength-1.0
                    else
                        value=2.0*(1.0-random_value)+2.0*(random_value-0.5)*(1.0-delta_2)^(1.0+eta)
                        delta_q=1.0-value^mutation_strength
                    end
                    mut[i,j]=mut[i,j]+delta_q*(1-0)
                    mut[i,j]=Base.min(Base.max(mut[i,j],0),1)
                elseif j>1
                    delta_1=(mut[i,j]-lower)/(upper-lower)
                    delta_2=(upper-mut[i,j])/(upper-lower)
                    mutation_strength=1.0/(1.0+eta)
                    random_value=Base.rand()
                    if random_value <= 0.5
                        value=2.0*random_value +(1.0-2.0*random_value)*(1.0-delta_1)^(1.0+eta)
                        delta_q=value^mutation_strength-1.0
                    else
                        value=2.0*(1.0-random_value)+2.0*(random_value-0.5)*(1.0-delta_2)^(1.0+eta)
                        delta_q=1.0-value^mutation_strength
                    end
                    mut[i,j]=mut[i,j]+delta_q*(upper-lower)
                    mut[i,j]=Base.min(Base.max(mut[i,j],lower),upper)
                end
            end
        end
    end
    return mut
end
        
    
```

```julia
function simulated_binary_crossover(ind1,ind2,eta,lower,upper)
    for k in 1:size(ind1,1)
        if Base.rand()<=prob_of_crossover
            if abs(ind1[k] - ind2[k]) > 0
                x1 = min(ind1[k], ind2[k])
                x2 = max(ind1[k], ind2[k])
                rand = Base.rand()
                beta = 1.0 + (2.0 * (x1 - lower) / (x2 - x1))
                alpha = 2.0 - beta^-(eta + 1)
                if rand <= 1.0 / alpha
                    beta_q = (rand * alpha)^(1.0 / (eta + 1))
                else
                    beta_q = (1.0 / (2.0 - rand * alpha))^(1.0 / (eta + 1))
                end
                c1 = 0.5 * (x1 + x2 - beta_q * (x2 - x1))
                beta = 1.0 + (2.0 * (upper - x2) / (x2 - x1))
                alpha = 2.0 - beta^-(eta + 1)
                if rand <= 1.0 / alpha
                    beta_q = (rand * alpha)^(1.0 / (eta + 1))
                else
                    beta_q = (1.0 / (2.0 - rand * alpha))^(1.0 / (eta + 1))
                end
                c2 = 0.5 * (x1 + x2 + beta_q * (x2 - x1))
                c1 = min(max(c1, lower), upper)
                c2 = min(max(c2, lower), upper)
                if Base.rand() <= 0.5   
                    ind1[k] = c2
                    ind2[k] = c1
                else
                    ind1[k] = c1
                    ind2[k] = c2
                end
            end
        end
    end
    return ind1,ind2
end
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
```

```julia
function simulated_binary_crossover_zdt_4(ind1,ind2,eta,lower,upper)
    for k in 1:size(ind1,1)
        if Base.rand()<=prob_of_crossover
            if k==1
                if abs(ind1[k] - ind2[k]) > 0
                    x1 = min(ind1[k], ind2[k])
                    x2 = max(ind1[k], ind2[k])
                    rand = Base.rand()
                    beta = 1.0 + (2.0 * (x1 - 0) / (x2 - x1))
                    alpha = 2.0 - beta^-(eta + 1)
                    if rand <= 1.0 / alpha
                        beta_q = (rand * alpha)^(1.0 / (eta + 1))
                    else
                        beta_q = (1.0 / (2.0 - rand * alpha))^(1.0 / (eta + 1))
                    end
                    c1 = 0.5 * (x1 + x2 - beta_q * (x2 - x1))
                    beta = 1.0 + (2.0 * (1 - x2) / (x2 - x1))
                    alpha = 2.0 - beta^-(eta + 1)
                    if rand <= 1.0 / alpha
                        beta_q = (rand * alpha)^(1.0 / (eta + 1))
                    else
                        beta_q = (1.0 / (2.0 - rand * alpha))^(1.0 / (eta + 1))
                    end
                    c2 = 0.5 * (x1 + x2 + beta_q * (x2 - x1))
                    c1 = min(max(c1, 0), 1)
                    c2 = min(max(c2, 0), 1)
                    if Base.rand() <= 0.5   
                        ind1[k] = c2
                        ind2[k] = c1
                    else
                        ind1[k] = c1
                        ind2[k] = c2
                    end
                end
            elseif k>1
                if abs(ind1[k] - ind2[k]) > 0
                    x1 = min(ind1[k], ind2[k])
                    x2 = max(ind1[k], ind2[k])
                    rand = Base.rand()
                    beta = 1.0 + (2.0 * (x1 - lower) / (x2 - x1))
                    alpha = 2.0 - beta^-(eta + 1)
                    if rand <= 1.0 / alpha
                        beta_q = (rand * alpha)^(1.0 / (eta + 1))
                    else
                        beta_q = (1.0 / (2.0 - rand * alpha))^(1.0 / (eta + 1))
                    end
                    c1 = 0.5 * (x1 + x2 - beta_q * (x2 - x1))
                    beta = 1.0 + (2.0 * (upper - x2) / (x2 - x1))
                    alpha = 2.0 - beta^-(eta + 1)
                    if rand <= 1.0 / alpha
                        beta_q = (rand * alpha)^(1.0 / (eta + 1))
                    else
                        beta_q = (1.0 / (2.0 - rand * alpha))^(1.0 / (eta + 1))
                    end
                    c2 = 0.5 * (x1 + x2 + beta_q * (x2 - x1))
                    c1 = min(max(c1, lower), upper)
                    c2 = min(max(c2, lower), upper)
                    if Base.rand() <= 0.5   
                        ind1[k] = c2
                        ind2[k] = c1
                    else
                        ind1[k] = c1
                        ind2[k] = c2
                    end
                end
            end
        end
    end
    return ind1,ind2
end
```

```julia
function local_reachability_density(initial_pop_decision_val)
    k=sqrt(size(initial_pop_decision_val,1))
    k=trunc(Int64,k)
    println("k value is")
    println(k)
    local_reachability_density=[]
    average_reachability_distance=[]
    dist_list_for_decision,sorted_dist_list,index_set_for_decision,kth_array = pairwise_distances_ties(initial_pop_decision_val,k)
#     println("index set is")
#     println(index_set_for_decision)
#     println("kth array is")
#     println(kth_array)
    for i in 1:size(index_set_for_decision,1)
        temp_sum_set=[]
        pushed_val=1
        for j in 1:size(index_set_for_decision,2)
            if i!=index_set_for_decision[i,j] && pushed_val<=k
#                 println(initial_pop_decision_val[i,:])
#                 println(initial_pop_decision_val[index_set_for_decision[i,j],:])
                dist_val=sqrt(reduce(+,(initial_pop_decision_val[i,:]-initial_pop_decision_val[index_set_for_decision[i,j],:]).^2))
#                 println("dist val is")
#                 println(dist_val)
                reachability_distance=max(kth_array[index_set_for_decision[i,j]],dist_val)
                push!(temp_sum_set,reachability_distance)
                pushed_val=pushed_val+1
            end 
        end 
        average_local_reachability_distance_val=sum(temp_sum_set)/k
        push!(average_reachability_distance,average_local_reachability_distance_val)
        local_reachability_density_val=1/(average_local_reachability_distance_val+2)
        push!(local_reachability_density,local_reachability_density_val)
    end
    
    return local_reachability_density,index_set_for_decision,sorted_dist_list,average_reachability_distance
end


    

#     reachability_distance(a,b)=max(kth distance of b, dist of a to b)
    
```

```julia
function local_outlier_factor(lrd,index_set,initial_pop_decision_val)
    k=sqrt(size(initial_pop_decision_val,1))
    k=trunc(Int64,k)
    lof=[]
    for i in 1:size(index_set,1)
        lrd_sum_val=0
        pushed_val=1
        for j in 1:size(index_set,2)
            if i!=index_set[i,j] && pushed_val<=k
                lrd_sum_val+=lrd[index_set[i,j]]
                pushed_val=pushed_val+1
            end
        end
        lof_val=(lrd[i]/(lrd_sum_val/k))
        push!(lof,lof_val)
    end
    return lof
end

                
        
    
```

```julia
function multimodal_analysis(initial_pop_decision_val,initial_pop_fitness_val)
    lrd,index_set,sdl,ard=local_reachability_density(initial_pop_decision_val)
#     println("local reachability density is")
#     println(lrd)
    tie_breaker=true
    lof=local_outlier_factor(lrd,index_set,initial_pop_decision_val)
#     println("local outlier factor is")
#     println(lof)
    for i in 1:size(initial_pop_fitness_val,1)
        key=initial_pop_fitness_val[i,:]
        for j in 1:size(initial_pop_fitness_val,1)
            if i!=j
                if key==initial_pop_fitness_val[j,:]
                    println("true")
                    tie_breaker=false
                    if lof[i]>lof[j]
                        initial_pop_fitness_val=initial_pop_fitness_val[1:end.!=j,:]
                        initial_pop_decision_val=initial_pop_decision_val[1:end.!=j,:]
                        if tie_breaker==false
                            return multimodal_analysis(initial_pop_decision_val,initial_pop_fitness_val)
                        end
                        
                    elseif lof[i]<lof[j]
                        
                        initial_pop_fitness_val=initial_pop_fitness_val[1:end.!=i,:]
                        initial_pop_decision_val=initial_pop_decision_val[1:end.!=i,:]
                        if tie_breaker==false
                            return multimodal_analysis(initial_pop_decision_val,initial_pop_fitness_val)
                        end
                    elseif lof[i]==lof[j]
                        temp=[]
                        push!(temp,i)
                        push!(temp,j)
                        chosen_val=StatsBase.sample(temp)
                        initial_pop_fitness_val=initial_pop_fitness_val[1:end.!=chosen_val,:]
                        initial_pop_decision_val=initial_pop_decision_val[1:end.!=chosen_val,:]
                        if tie_breaker==false
                            return multimodal_analysis(initial_pop_decision_val,initial_pop_fitness_val)
                        end 
                    end
                end
            end
        end
    end
    
    return initial_pop_decision_val,initial_pop_fitness_val
end
    


    

    
```

```julia
function archive_truncation(non_set,fitness_set,solution_set)
    setup=[val < 1 for val in non_set]
    val=count(c->c==true,setup)
    println("total number of non dominated solutions found(which means raw_fitness < 1)")
    println(val)
    if val < archive_size
        index=sortperm(non_set)
        fitness_set=fitness_set[index,:]
        solution_set=solution_set[index,:]
        non_set=non_set[index]
        fitness_set=fitness_set[1:archive_size,:]
        solution_set=solution_set[1:archive_size,:]
        non_set=non_set[1:archive_size]
    elseif val==archive_size
        index=sortperm(non_set)
        fitness_set=fitness_set[index,:]
        solution_set=solution_set[index,:]
        non_set=non_set[index]
        fitness_set=fitness_set[1:archive_size,:]
        solution_set=solution_set[1:archive_size,:]
        non_set=non_set[1:archive_size]
    elseif val > archive_size
        index_list=[]
        for (index,val) in enumerate(non_set)
            if val<1
                push!(index_list,index)
            end
        end
        fitness_set=fitness_set[index_list,:]
        solution_set=solution_set[index_list,:]
        non_set=non_set[index_list]
        counter_setup=size(fitness_set,1)-archive_size
        solution_set,fitness_set,non_set=truncation_setup(non_set,fitness_set,solution_set,counter_setup)
    end
    return solution_set,fitness_set,non_set
end
    

      
    
    
    
    
```

```julia
function pairwise_distances(pop)
    k::Int64=floor(sqrt(size(pop,1)))
    dist_list=Matrix(undef,size(pop,1),size(pop,1))
    index_dist_list=Matrix(undef,size(pop,1),size(pop,1))
    for i in 1:size(pop,1)
        key=pop[i,:]
        for j in 1:size(pop,1)
            val=(key-pop[j,:]).^2
            euclidean_distance=reduce(+,val)
            dist_list[i,j]=(sqrt(euclidean_distance))
        end
        index=sortperm(dist_list[i,:])
        index_dist_list[i,:]=index
        val=sort(dist_list[i,:])
        dist_list[i,:]=val
    end
    kth_array=dist_list[:,k+1]
    return dist_list,index_dist_list,kth_array,k
end




```

```julia
function pairwise_distances_ties(pop,k)
    
    println("fitness size is")
    println(size(pop,1))
    
    dist_list=Matrix(undef,size(pop,1),size(pop,1))
    index_dist_list=Matrix(undef,size(pop,1),size(pop,1))
    sorted_dist_list=Matrix(undef,size(pop,1),size(pop,1))
    kth_array=[]
    val_checker=[]
    for i in 1:2
        push!(val_checker,-1)
    end

    for i in 1:size(pop,1)
        key=pop[i,:]
        for j in 1:size(pop,1)
            if pop[j,:]==val_checker || pop[i,:]==val_checker
                dist_list[i,j]=-1.0
            elseif pop[j,:]!=val_checker && pop[i,:]!=val_checker
                val=(key-pop[j,:]).^2
                euclidean_distance=reduce(+,val)
                dist_list[i,j]=sqrt(euclidean_distance)
            end
        end
        index=sortperm(dist_list[i,:])
        index_dist_list[i,:]=index
        val=sort(dist_list[i,:])
        sorted_dist_list[i,:]=val
    end
    
    for i in 1:size(dist_list,1)
        count=0
        neighbor=sort(dist_list[i,:])
        for (inner_index,temp_val) in enumerate(neighbor)
    
            if temp_val>0
                count=count+1
            end
            if count==k
                push!(kth_array,neighbor[inner_index])
            end
        end
    end
    
    return dist_list,sorted_dist_list,index_dist_list,kth_array
end


```

```julia
function binary_tournament_selection(pop,fit_pop,non_set)
    offspring_size::Int64=size(non_set,1)/2
    potential_parents=rand(100,decision_space_dimensions)
    temp=collect(1:size(pop,1))
    lrd,index_set,sorted_dist_list,kth_array=local_reachability_density(fit_pop)

    for i in 1:100
        @label start_loop
        val_one=StatsBase.sample(temp)
        val_two=StatsBase.sample(temp)
        if val_one!=val_two   
            if non_set[val_one]>non_set[val_two]
                potential_parents[i,:]=pop[val_two,:]
#                 push!(potential_parents,pop[val_two,:])
            elseif non_set[val_one]<non_set[val_two]
                potential_parents[i,:]=pop[val_one,:]
#                 push!(potential_parents,pop[val_one,:])
            elseif non_set[val_one]==non_set[val_two]
                if kth_array[val_one]>kth_array[val_two]
                    potential_parents[i,:]=pop[val_one,:]
                elseif kth_array[val_one]<kth_array[val_two]
                    potential_parents[i,:]=pop[val_two,:]
                elseif kth_array[val_one]==kth_array[val_two]
                    @goto start_loop
                end
            end
        elseif val_one==val_two
            @goto start_loop
        end
    end
    return potential_parents
end


        
        
        
        
    
```

```julia
function weakly_non_domination_verification(fitness_set,density_set)
#     non dominated setup
    fitness_obj_1=Array{Float64}(undef,size(fitness_set,1))
    fitness_obj_2=Array{Float64}(undef,size(fitness_set,1))
    dominators=[]
    list_of_dominations=[]
    raw_fitness=Array{Float64}(undef,size(fitness_set,1))
    for i in 1:size(fitness_set,1)
        fitness_obj_1[i]=fitness_set[i,1]
        fitness_obj_2[i]=fitness_set[i,2]
    end
    for i in 1:size(fitness_obj_1,1)
        dominator=0
        temp_list=[]
        for j in 1:size(fitness_obj_2,1)
            if (fitness_obj_1[i] < fitness_obj_1[j] && fitness_obj_2[i] < fitness_obj_2[j]) || (fitness_obj_1[i] <= fitness_obj_1[j] && fitness_obj_2[i] < fitness_obj_2[j]) || (fitness_obj_1[i] < fitness_obj_1[j] && fitness_obj_2[i] <= fitness_obj_2[j]) 
                dominator+=1
            elseif (fitness_obj_1[i] > fitness_obj_1[j] && fitness_obj_2[i] > fitness_obj_2[j]) || (fitness_obj_1[i] >= fitness_obj_1[j] && fitness_obj_2[i] > fitness_obj_2[j]) || (fitness_obj_1[i] > fitness_obj_1[j] && fitness_obj_2[i] >= fitness_obj_2[j])
                push!(temp_list,j)
            end
        end
        push!(dominators,temp_list)
        push!(list_of_dominations,dominator)
    end
    
    #raw_fitness_setup
    raw_fitness_setup=[]
    for (i,val) in enumerate(dominators)
        raw_fitness=0
        if length(val)!=0
            for j in 1:size(val,1)
                raw_fitness+=list_of_dominations[val[j]]
            end
#         println("raw fitness before adding density val",raw_fitness)
            raw_fitness+=density_set[i]
#         println("raw fitness after adding density val",raw_fitness)
            push!(raw_fitness_setup,raw_fitness)
        elseif length(val)==0
             raw_fitness=density_set[i]
             push!(raw_fitness_setup,raw_fitness)
        end
    end
    return raw_fitness_setup
end
 
# non_dom=non_domination_verification(fit,dense)
# println(non_dom)
```

```julia
function igd_indicator(true_pareto_front,obtained_pareto_front)
    dist_euclidean=[]
    sum_min=0
    for val_one in true_pareto_front
        for val_two in obtained_pareto_front
            d=sqrt(reduce(+,((val_one-val_two).^2)))
            push!(dist_euclidean,d)
        end
        dist_euclidean=sort(dist_euclidean,rev=true)
        sum_min+=pop!(dist_euclidean)
        dist_euclidean=[]
    end
    igdval=sum_min/len(true_pareto_front)
    return igdval
end

            
```

```julia
import Statistics

function schott_spacing_metric(obtained_pareto_front)
    val_k::Int64=floor(sqrt(size(obtained_pareto_front,1)))
    dist_list,sorted_dist_list,index_set,kth_array=pairwise_distances_ties(obtained_pareto_front,val_k)
    dist_array=[]
    for i in 1:size(index_set,1)
        dist=sqrt(reduce(+,(obtained_pareto_front[index_set[i,1],:]-obtained_pareto_front[index_set[i,2],:]).^2))
        push!(dist_array,dist)
    end
    
    mean_dist=Statistics.mean(dist_array)
    diff=(mean_dist .-dist_array)
    diff=diff.^2
    
    
    schott_val=sqrt(sum(diff)/(size(obtained_pareto_front,1)-1))
    
    return schott_val
end
    
    
```

```julia
function joining_matrix_after_modifications(mat_a,mat_b)
#     mat_a=Base.transpose(mat_a)
#     mat_b=Base.transpose(mat_b)
    
    
    mat_a=transpose(mat_a)
    mat_b=transpose(mat_b)
    
    
    row_of_mat_a=size(mat_a,1)
#     println(row_of_mat_a)
    col_of_mat_a=size(mat_a,2)
#     println(col_of_mat_a)
    row_of_mat_b=size(mat_b,1)
#     println(row_of_mat_b)
    col_of_mat_b=size(mat_b,2)
#     println(col_of_mat_b)
    reshaped_mat_a=reshape(mat_a,((row_of_mat_a*col_of_mat_a),1))
    reshaped_mat_b=reshape(mat_b,((row_of_mat_b*col_of_mat_b),1))
#     println("reshaped matrices")
#     println(reshaped_mat_a)
#     println(reshaped_mat_b)
    setup=vcat(reshaped_mat_a,reshaped_mat_b)
#     println("matrix after vcat")
#     println(setup)
    size_of_setup=size(setup,1)
#     println(size_of_setup)
    col_of_final_mat::Int64=size_of_setup/decision_space_dimensions
#     println(col_of_final_mat)
    final_mat=reshape(setup,(decision_space_dimensions,col_of_final_mat))
    
    return transpose(final_mat)
 end
```
